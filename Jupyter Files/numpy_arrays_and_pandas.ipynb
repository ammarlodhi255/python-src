```python
import numpy as np

my_arr = np.array([[1, 2, 3, 4, 5.5], [9.3, 5.2, 8.7, 9, 1]], dtype='f')
print(f"my_arr: {my_arr}\nShape: {my_arr.shape}\nndim: {my_arr.ndim}")

for i, elem in enumerate(np.nditer(my_arr)):
    print(f"elem {i}: {elem}")

my_arr = np.array([2, 3, 4, 5])
new_arr = np.array(np.array_split(my_arr, 3))
print(f"split arr: {new_arr}")
```
